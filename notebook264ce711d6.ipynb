```python
import requests
from bs4 import BeautifulSoup

def get_survey_numbers(district, mandal, village):
    url = "https://dharani.telangana.gov.in/knowLandStatus"
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
    }
    data = {
        "districtCode": district,
        "mandalCode": mandal,
        "villageCode": village
    }
    try:
        response = requests.post(url, headers=headers, data=data, timeout=30)  # Increase timeout to 30 seconds
        response.raise_for_status()  # Raise an exception for 4XX or 5XX status codes
        soup = BeautifulSoup(response.content, 'html.parser')
        survey_numbers = []
        for option in soup.find_all('option'):
            survey_numbers.append(option.text.strip())
        return survey_numbers[1:]  # Skipping the first option which is "--Select--"
    except requests.exceptions.RequestException as e:
        print("An error occurred:", e)
        return []

district_code = input("Enter district code: ")
mandal_code = input("Enter mandal code: ")
village_code = input("Enter village code: ")

survey_numbers = get_survey_numbers(district_code, mandal_code, village_code)
if survey_numbers:
    print("Survey numbers for the selected district, mandal, and village are:")
    for survey_number in survey_numbers:
        print(survey_number)
else:
    print("No survey numbers found for the selected district, mandal, and village.")
```
